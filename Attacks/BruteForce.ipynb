```python
def card_decrypt(d_key, c2):
        val = int(c2)
        d1 = pow(val,d_key,n)
        byt_val = long_to_bytes(d1)
        msg = byt_val.decode()
        return msg
```

```python
results=[]
for d_key in possible_d:
  for value in deck:
    try:
     # print(int(d_key))
      results.append(card_decrypt(int(d_key),value))
    except:
      continue
  if (results[0].isdigit() and results[1].isdigit()):
    print(results)
    print("Decryption key: ", d_key)
  else:
    results =[]
```
